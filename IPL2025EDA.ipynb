## 18. standardizing column names
This operation cleans and standardizes column names by removing leading and trailing spaces,converting all names to lowercase,and replacing spaces with underscores.It ensures consistency,avoids naming conflicts,and makes columns easier to reference during analysis and coding.

```python
batters_df.columns = (
    batters_df.columns
    .str.strip()
    .str.lower()
    .str.replace('*','',regex = False)
)
```

## 19. converting columns to Numeric Data types.
This operation converts the avg and sr columns in both batters and bowlers datasets into numeric format. Any invalid or non-numeric values are coerced into NaN, ensuring accurate calculations and reliable statistical analysis.


```python
batters_df['avg'] = pd.to_numeric(batters_df['avg'],
errors='coerce'
)

batters_df['sr'] = pd.to_numeric(batters_df['sr'],                            
errors='coerce'
)
```

## 23. Top 5 Players with the Highest Run Scores
This section presents a comparative analysis of the top five run-scorers against the season average.

```python
batters_df = pd.read_csv('cleaned_batters.csv')
highest_run_scorer = batters_df[['player name','runs']].sort_values(by='runs',ascending=False).head(5)
highest_run_scorer
```

```python
plt.bar(x = highest_run_scorer['player name'],height = highest_run_scorer['runs'],color=['#9B7FD3'])

plt.show()
```

**Insight**:    
Top run-scorers consistently outperform the season average, underscoring the influence of 
elite batting performances.

## 24. Top 5 Teams by Total Wickets Taken
This section provides a comparative assessment of the five leading teams based on total wickets taken.

```python
bowlers_df=pd.read_csv('cleaned_bowlers.csv')
team_grp_df = bowlers_df.groupby('team')
df = team_grp_df['wkt'].sum().sort_values(ascending=False)
df.head(5)
```

```python
plt.bar(x=df.index,height=df.values,color=["#AB0AE6"])
plt.xlabel('Teams')
plt.ylabel('No of Wickets')
plt.title('Most wickets by each Team')
plt.show()
```

**Insight**:     
Superior bowling performance is reflected in higher wicket 
totals, 
offering teams a decisive competitive edge.

## 25. Top 5 Power Hitters: High Strike Rate, Most Sixes (10+ Matches)

## 26. Run-Scoring Giants: Teams Exceeding 2500 Runs.
This section showcases teams that have crossed the 2500-run milestone, reflecting sustained dominance with the bat.

```python
team_grp_df = batters_df.groupby('team')
df = team_grp_df['runs'].sum().sort_values(ascending=False)
df[df.values>2500]
```

```python
s = batters_df.groupby('team')['runs'].sum()
s = s[s > 2500]

plt.plot(s.index, s.values, marker='o')
plt.show()
```

**Insight**:  
Crossing the 2500-run mark signals superior batting depth, consistency, and offensive control throughout the season.

## 27. Elite All-Rounders Showdown: Runs, Wickets, and Total Impact.
This section presents a high-level comparative analysis of the top six all-rounders, measuring batting output, bowling effectiveness, and overall match impact.


```python
allrounders = pd.merge(
    batters_df[["player name", 
"runs"]],
    bowlers_df[["player name", 
"wkt"]],
    on="player name",
    how="inner"
)

allrounders = allrounders[
    (allrounders["runs"] > 0) &
    (allrounders["wkt"] > 0)
]

allrounders["Performance"] = allrounders["runs"] + allrounders["wkt"]

top5_allrounders = (
    allrounders
    .sort_values(by="Performance", ascending=False)
    .head(5)
)

print(top5_allrounders)
```

```python
top5_allrounders.set_index("player name")[["runs", "wkt"]].plot(kind="bar")
plt.xticks(rotation=0,ha="right")
plt.xlabel("player name")
plt.ylabel("Count")
plt.title("Top 5 All-Rounders: Runs vs Wickets")
plt.show()
```

**Insight**:    
The most dominant all-rounders distinguish themselves by excelling simultaneously in run accumulation and wicket-taking, maximizing total influence on results.
